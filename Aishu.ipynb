```python
import matplotlib.pyplot as plt
```

```python
people=['Kiran','Arun','Viay','Varun']
```

```python
age=[25,30,35,40]
```

```python
height=[145,151,165,173]
```

```python
weight=[45,55,65,75]
```

```python
plt.figure(figsize=(5,5))
```

```python
plt.scatter(age,height,s=weight,c='blue',alpha=0.7,label='Weight')
```

```python
plt.title('Scatter Plot of Age vs Height with Weight')
```

```python
plt.xlabel('Age')
```

```python
plt.ylabel('Height')
```

```python
plt.legend()
```

```python
plt.grid(True)
```

```python
plt.show()
```

```python
plt.figure(figsize=(8,6))
```

```python
bar_width=0.3
```

```python
index=range(len(people))
```

```python
plt.bar(index,weight,bar_width,color='green',label='weight')
```

```python
plt.title('Bar Chart of Weight for Each Person')
```

```python
plt.xlabel('People')
```

```python
plt.ylabel('Weight')
```

```python
plt.xticks(index,people)
```

```python
plt.legend()
```

```python
plt.show()
```

```python
plt.figure(figsize=(8,6))
```

```python
plt.hist(age,bins=5,color='orange',edgecolor='black')
```

```python
plt.title('Histogram of Age Distribution')
```

```python
plt.xlabel('Age')
```

```python
plt.ylabel('Frequency')
```

```python
plt.grid(axis='y',alpha=0.75)
```
